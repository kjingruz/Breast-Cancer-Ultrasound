```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt

def load_data(data_path, image_size=(128, 128)):
    images = []
    masks = []
    subfolders = ['benign', 'malignant', 'normal']  # Specify subdirectories
    for subfolder in subfolders:
        full_path = os.path.join(data_path, subfolder)
        for filename in os.listdir(full_path):
            if '_mask' in filename:
                continue
            img_path = os.path.join(full_path, filename)
            mask_path = img_path.replace('.png', '_mask.png')
            
            img = tf.keras.preprocessing.image.load_img(img_path, target_size=image_size, color_mode="grayscale")
            mask = tf.keras.preprocessing.image.load_img(mask_path, target_size=image_size, color_mode="grayscale")
            
            img = tf.keras.preprocessing.image.img_to_array(img)
            mask = tf.keras.preprocessing.image.img_to_array(mask)
            mask = mask / 255.0  # Normalize mask to [0, 1]
            
            images.append(img)
            masks.append(mask)
    
    return np.array(images), np.array(masks)


from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

def get_unet(input_size=(128,128,1)):
    inputs = Input(input_size)
    # Encoder
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    conv1 = Dropout(0.2)(conv1)
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
    
    # Bottleneck
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1)
    conv2 = Dropout(0.2)(conv2)
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    # Decoder
    up1 = UpSampling2D(size=(2, 2))(pool2)
    conv3 = Conv2D(64, (3, 3), activation='relu', padding='same')(up1)
    conv3 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv3)

    up2 = UpSampling2D(size=(2, 2))(conv3)
    # Ensure that the concatenation is with a corresponding layer of the same size
    concat1 = concatenate([up2, conv1], axis=3)

    conv4 = Conv2D(32, (3, 3), activation='relu', padding='same')(concat1)
    conv4 = Dropout(0.2)(conv4)
    conv4 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv4)

    # Output layer
    output = Conv2D(1, (1, 1), activation='sigmoid')(conv4)

    model = Model(inputs=inputs, outputs=output)
    # Updated Adam optimizer instantiation
    model.compile(optimizer=Adam(learning_rate=1e-4), loss='binary_crossentropy', metrics=['accuracy'])
    return model



# Main workflow
data_path = './'  # Adjust this path
images, masks = load_data(data_path)
x_train, x_val, y_train, y_val = train_test_split(images, masks, test_size=0.1, random_state=42)

model = get_unet()
model.summary()
model.fit(x_train, y_train, batch_size=2, epochs=50, verbose=1, validation_data=(x_val, y_val))

# Prediction
sample_test = x_val[:5]
preds = model.predict(sample_test, verbose=1)
preds_t = (preds > 0.5).astype(np.uint8)

# Display results
fig, axs = plt.subplots(5, 2, figsize=(10, 25))
for i in range(5):
    axs[i, 0].imshow(sample_test[i].squeeze(), cmap='gray')
    axs[i, 0].title.set_text('Original Image')
    axs[i, 1].imshow(preds_t[i].squeeze(), cmap='gray')
    axs[i, 1].title.set_text('Predicted Mask')
plt.show()
```
